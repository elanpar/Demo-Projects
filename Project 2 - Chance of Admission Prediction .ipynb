```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns
import pickle

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('D:\Python\Data\Admission_Prediction.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
# Fill the null values

data ['University Rating'] = data['University Rating'].fillna(data['University Rating'].mode()[0])

data ['GRE Score'] = data['GRE Score'].fillna(data['GRE Score'].mean())

data ['TOEFL Score'] = data['TOEFL Score'].fillna(data['TOEFL Score'].mean())
```

```python
data.isna().sum()
```

```python
data = data.drop(columns = ['Serial No.'])
```

```python
plt.figure(figsize=(20,15),facecolor = 'Yellow')
plotnumber = 1

for column in data:
    if plotnumber <= 8:
        ax = plt.subplot(2,4,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize = 20)
        
    plotnumber += 1
    
plt.show() or plt.tight_layout()
```

```python
y = data['Chance of Admit']
x = data.drop(columns = ['Chance of Admit'])
```

```python
## EDA = Exploratory Data Analysis

plt.figure(figsize=( 20,10),facecolor = 'yellow')

plotnumber = 1

for column in x:
    if plotnumber<=8:
        ax = plt.subplot(2,4,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column, fontsize = 10)
        plt.ylabel('Chance of Admit',  fontsize = 10)
        
    plotnumber += 1
plt.show()
```

```python
## Data Scalling  formula z = (x - mean)/std

scaler = StandardScaler()
X_Scaled = scaler.fit_transform(x)

X_Scaled
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_Scaled,y,test_size = 0.25, random_state=49)
y_train.head()
```

```python
## Model Instantiating and training

regression = LinearRegression()
regression.fit(x_train, y_train)
```

```python
data.tail(2)
```

```python
# Predict the chance of admission

print('The chance of Admission is = ',regression.predict(scaler.transform([[312.0,113.0,4.5,4.5,5.0,9.04,0]])))
```

```python
# saving the model in the local file system

filename = 'finalized_model.pickle'
pickle.dump(regression, open(filename,'wb'))


## Prediction using the saved model

loaded_model = pickle.load(open(filename, 'rb'))

a = loaded_model.predict(scaler.transform([[314,103,2,2,3,8.21,0]]))
a
```

```python
## Adjusted R2

regression.score(x_train,y_train)
```

```python
regression.score(x_test,y_test)
```

```python
## Lets plot and visualize

y_pred = regression.predict(x_test)
y_pred
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Chance of Admission')
plt.ylabel('Predicted Chance of Admission')
plt.title('Actual Vs Model Predicted')
plt.show()
```

```python
## Model Evaluation

from sklearn.metrics import mean_squared_error, mean_absolute_error

y_pred = regression.predict(x_test)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
## Reularization


from sklearn.linear_model import Ridge,Lasso,RidgeCV, LassoCV

lasscv = LassoCV(alphas = None, max_iter = 100, normalize = True)
lasscv.fit(x_train,y_train)
```

```python
# Best alpha parameter

alpha = lasscv.alpha_
alpha
```

```python
## Lasso Regression

lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
ridgecv = RidgeCV(alphas = np.arange(0.001,0.1,0.01), normalize = True)

ridgecv.fit(x_train,y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_model = Ridge(alpha = ridgecv.alpha_)
ridge_model.fit(x_train,y_train)

ridge_model.score(x_test,y_test)
```
